```python
### 각 연도별 매출액 데이터에서 필요 없는 행 열 제거 후

def preprocessing(*yeardatas):
    for yeardata in yeardatas:
        data = pd.read_csv('/Users/hanjaemin/Desktop/seoul_data/original_data/상권/상권_추정매출액/'+yeardata+'.csv',encoding = 'euc-kr')
        data.drop(columns = ['서비스_업종_코드'], axis = 1, inplace = True)
        data.drop(columns = ['상권_코드'], axis = 1, inplace = True)
        data.drop(columns = ['상권_코드_명'], axis = 1, inplace = True)
#         data.drop(columns = ['서비스_업종_코드_명'], axis = 1, inplace = True)
        data.drop(columns = ['상권_구분_코드_명'], axis = 1, inplace = True)
        data.to_csv('/Users/hanjaemin/Desktop/seoul_data/merge_data/' + yeardata + '.csv')
```
